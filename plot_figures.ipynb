```python
import matplotlib.pyplot as plt
import pickle
```

```python
logger_et_nl = pickle.load(open("./logger_et_nl.pkl", "rb"))
logger_tb_nl = pickle.load(open("./logger_tb_nl.pkl", "rb"))
states_et_nl = list(zip(*logger_et_nl['states']))
controls_et_nl = list(zip(*logger_et_nl['controls']))
states_tb_nl = list(zip(*logger_tb_nl['states']))
controls_tb_nl = list(zip(*logger_tb_nl['controls']))
interval_et_nl = list(zip(*logger_et_nl['interval_et']))


font_size = 18
label_size = 18
x_lim_nl = 1



plt.close('all')
plt.style.use('seaborn-talk')
```

```python
fig = plt.figure()

##
plt.subplot(2,1,1)
plt.plot(logger_et_nl['steps'], states_et_nl[2])
plt.plot(logger_tb_nl['steps'], states_tb_nl[2], linestyle='--')

# plt.xlabel('Time [s]',fontsize = font_size)
plt.ylabel(r'$\delta_e~\mathrm{[deg]}$',fontsize = font_size)
plt.xticks(fontsize = font_size)
plt.yticks(fontsize = font_size)
plt.xlim(0, x_lim_nl)
# plt.ylim(-8, 4)
plt.grid(color='gray', linestyle=':')
plt.legend(["Event-triggered DHP", "Time-based DHP"], loc='lower right',fontsize = font_size)

##
plt.subplot(2,1,2)
plt.plot(logger_et_nl['steps'], controls_et_nl[0])
plt.plot(logger_tb_nl['steps'], controls_tb_nl[0], linestyle='--')

plt.xlabel('Time [s]',fontsize = font_size)
plt.ylabel(r'$\delta_e^c~\mathrm{[deg]}$',fontsize = font_size)
plt.xticks(fontsize = font_size)
plt.yticks(fontsize = font_size)
plt.xlim(0, x_lim_nl)
# plt.ylim(-15, 12)
plt.grid(axis='both',color='gray', linestyle=':')



plt.tight_layout()
plt.savefig('./figures/state_1_nl.pdf', format='pdf', dpi=1000, bbox_inches = 'tight')
```
